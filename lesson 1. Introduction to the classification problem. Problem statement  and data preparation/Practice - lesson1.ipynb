Вопросы по теоретической части урока 1:

### 1. Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?
    
- __precision__ = $\cfrac{TP}{TP+FP}$ - коэффициент, который показывает множество __верно__ предсказанных объектов из множества __всех предсказанных__ объектов. <br>

- __recall__ = $\cfrac{TP}{TP+FN}$ - коэффициент, который показывает множество __верно__ предсказанных объектов из множества __всех искомых__ объектов. <br>



Чем __выше__ коэффициент __precision__ тем меньше объектов __лишнего__ множества попадет в __предсказанную выборку__. При этом коэффициент __recall__ так же может упасть.<br>
Повышение коэффициента __precision__ важно если потери на __поиск__ объектов искомого множества выше потерь при __упускании__ объектов искомого множества:

    1) Высокая стоимость на распростарение рекламного контента
    2) Выдача большой суммы кредита

Чем __выше__ коэффициент __recall__ тем больше объектов __искомого__ множества попадет в __предсказанную выборку__. При этом коэффициент __precision__ так же может упасть.<br>
Повышение коэффициента __recall__ важно если потери при __упускании__ объектов искомого множества выше потерь на __поиск__ объектов искомого множетса:

    1) Предотвращение террористических актов
    2) Предотвращение аврийных ситуаций

### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

- Думаю эта метрика позволит нам настраивать её приоритет, при подсчете, между precision и recall, что не позволит сделать подсчет среднего значения.
