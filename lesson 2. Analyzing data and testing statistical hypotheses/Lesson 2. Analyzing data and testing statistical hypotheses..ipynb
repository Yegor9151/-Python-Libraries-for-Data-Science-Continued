### 1. В чём различие между зависимыми и независимыми выборками?

__Зависимые выборки__ - это те которые совпадают по признакам <br>
Пример:
- выборки одних и тех же людей до и после проведения эксперимента

__Не зависимые выборки__ - это те которые не совпадают по ризнакам<br>
Пример:
- выборка кошек и собак

### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

__Параметрические статисти__ - это те, которые используют какие параметры как __среднее значение, квантили, дисперсия и среднее квадратическое отклонение__ - эти параметры имеют выборки с __нормальным__ распределением, их график похож на Гаусовскую кривую.<br>
Следовательно если исследуемая выборка __имеет нормальное распределение__, тогда __параметрические статистические критерии__ будут иметь большую точность по сравнению с не параметрическими.

- t-критерий Стьюдента
- Критерий Фишера
- Критерий отношения правдоподобия
- Критерий Романовского

__Не параметрические статиститки__ - это те, которые используют параметры __отличные от__ параметримеских, их график не похож на Гаусовскую кривую.<br>
Следовательно если же исследуемая выборка __не имеет нормальное распределение__, тогда __не параметримеские статистические критерии__ будут иметь большую точность по сравнению с параметрическими.

- Q-критерий Розенбаума
- U-критерий Манна — Уитни
- Критерий Уилкоксона
- Критерий Пирсона
- Критерий Колмогорова — Смирнова
