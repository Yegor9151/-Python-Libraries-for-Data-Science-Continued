### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

Варианты усреднения для __f1_score__:

- __micro__ - считает среднее значение по сумме значенияй из категорий: TP, FN, FP.<br>
Следовательно используется для оченки влияния TN.

- __macro__ - считает по каждой категории (TP, FP, TN, FN), затем находит их не взвешенное среднее значение.<br>
При этом не учитывается дисбаланс предсказываемых классов.<br>
Следовательно используется если исследуемые классы имеют равное количество объектов.

- __weighted__ - считает по каждой категории (TP, FP, TN, FN), затем находит их взвешенное среднее значение по количеству истинных вариантов для каждого класса. Является альтернативой для "macro" при подсчете не сбалансированных классов. Это может привести к результату, который не находится между _точностью_ и _охватом_.<br>
Следовательно используется если исследуемые классы имеют не равное количество объектов.

### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

_XGBoost:_
- Достаточно старая модель, разработанная в 2014 году
- Склонен к быстрому переобучению
- Имеет большой количество гиперпараметров
- Имена некоторых гиперпараметров отличается от аналогичных в других моделях
- Использует алгоритм предварительной сортировки и алгоритм на основе гистограммы, для поиска наилучшего разбиения
- Не работает с категориальными и строковыми данными

_LightGBM:_
- Одна из новых моделей, разработанная в 2017 году
- Отличается высокой скоростью
- Использует алгоритм градиентного одностороннего разбиения - GOSS
- Умеет работать с категориальными признаками, 
- Нужно приводить признаки которые МЫ считаем категорийными к типу *int*.
- Не работает со строковыми типами
- Имеет очень важный для себя параметр - число листьев (num_leaves), по умолчанию имеет значение $2^{max\_depth}$
	
_СatBoost:_
- Одна из новых моделей, разработанная в 2017 году
- Разработана в российской компании Яндекс
- Умеет работать с категориальными и строковыми произнаками
- Нужно приводить признаки которые МЫ считаем категорийными к типу *category*, что бы алгоритм не рассматривал из как числовые
- Один из самых долгих алгоритмов
- Имеет графическое представление
- По умолчанию выдает информацию по каждой итерации
- Не имеет гиперпараметров для настройки подмножеств и минимального размера ответвления (min_child_weight)
